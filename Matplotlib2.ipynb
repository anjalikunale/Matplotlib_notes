### bar charts and analyzing data from CSVs

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.style.use("fivethirtyeight")

ages_x=list(range(25,36))
dev_y=[39000,40000,43000,46000,50000,55000,57000,60000,64000,66000,69000]
plt.plot(ages_x,dev_y,label="all dev")
plt.legend()
plt.show()



```

```python
plt.title("Median salary by ages")
plt.xlabel("Age")
plt.ylabel("Median salary")
plt.bar(ages_x,dev_y,label="All dev")
plt.legend()
plt.show()
```

```python
plt.title("Median salry by ages")
plt.xlabel("Ages")
plt.ylabel("Median salary")

py_dev_y=[40000,43000,47000,50000,53000,55000,59000,63000,67000,70000,73000]
js_dev_y=[30000,33000,37000,40000,43000,46000,49000,53000,56000,57000,60000]

plt.bar(ages_x,dev_y,label="all dev")
plt.bar(ages_x,py_dev_y,label="python dev")
plt.bar(ages_x,js_dev_y,label="js dev")

plt.legend()
plt.show()

```

```python
import numpy as np
x_indexes=np.arange(len(ages_x))

plt.bar(x_indexes,dev_y,label="all dev")
plt.bar(x_indexes,py_dev_y,label="python dev")
plt.bar(x_indexes,js_dev_y,label="js dev")

plt.legend()
plt.show()


```

```python
width=0.25
plt.title("Salary median by age")
plt.xlabel("Age")
plt.ylabel("Median salary")
plt.bar(x_indexes-width,dev_y,width=width,label="all dev")
plt.bar(x_indexes,py_dev_y,width=width,label="python dev")
plt.bar(x_indexes+width,js_dev_y,width=width,label="js dev")

plt.legend()
plt.show()


```

```python
html_dev_y=[30000,32000,34000,37000,38000,40000,43000,46000,50000,55000,58000]

plt.title("salary by age")
plt.xlabel("age")
plt.ylabel("salary")

plt.bar(x_indexes-width,dev_y,width=width,label="All dev")
plt.bar(x_indexes,py_dev_y,width=width,label="python dev")
plt.bar(x_indexes+width,js_dev_y,width=width,label="js dev")
plt.bar(x_indexes+width+width,html_dev_y,width=width,label="html dev")

plt.legend()
plt.show()
```

```python
#n above plot we seen that our 
#axis has indexes as label instead 
#of age, let's make it correct'''

plt.title("salary median by age")
plt.xlabel("Age")
plt.ylabel("Median salary")
plt.style.use("fivethirtyeight")
#plt.style.use("ggplot")


plt.bar(x_indexes-width,dev_y,width=width,label="all dev")
plt.bar(x_indexes,py_dev_y,width=width,label="python dev")
plt.bar(x_indexes+width,js_dev_y,width=width,label="js dev")

plt.legend()
plt.xticks(ticks=x_indexes,labels=ages_x)
plt.show()


```

```python
#access languages and it's popularity from csv file

import csv
with open("/storage/emulated/0/DS_PythonBasic/popularlanguage.csv","r") as f:
    r=csv.DictReader(f)
    language_count=Counter()
    for i in r:
        language_count.update(i["programminglanguage"].split(";"))

language=[]
popularity=[]
#print(language_count)
for l in language_count.most_common(10):
    language.append(l[0])
    popularity.append(l[1])
    
    '''use below code when you travel loop
    on language_count.most_common(3) 
    bcz it gives list of tuples'''
    #language.append(l[0])
    #popularity.append(l[1])

print(language,"\n", popularity)
print(language_count)   
    
    
    
```

```python
#above plot bar as horizontal plot bar

plt.title("popular programming language")
plt.ylabel("programming language")
plt.xlabel("no of student who use")

plt.barh(language, popularity)

plt.show()
```

```python
#plot in decreasing order of popularity

plt.title("popular programming language")
plt.xlabel("no of students who use")
plt.ylabel("programming language")
language.reverse()
popularity.reverse()
plt.barh(language, popularity)

plt.show()

```

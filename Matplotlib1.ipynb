```python
from matplotlib import pyplot as plt

ages_x=list(range(25,37))
salary_y=[38000,42000,46000,49000,52000,54000,62000,64000,67000,68000,70000,72000]
plt.plot(ages_x,salary_y)
```

```python
%matplotlib inline
plt.plot(ages_x,salary_y)
```

```python
plt.plot(ages_x,salary_y)
plt.show()
```

```python
salary2_y=[45000,48000,49000,51000,53000,57000,63000,65000,70000,71000,75000,83000]
plt.plot(ages_x,salary2_y)
plt.show()
```

```python
plt.title("Salary Data analysis")
plt.xlabel("Ages")
plt.ylabel("Salary median")
plt.plot(ages_x,salary_y)
plt.plot(ages_x,salary2_y)
plt.show()
```

```python
plt.title("Salary median by ages")
plt.xlabel("Ages")
plt.ylabel("Salary median")
plt.plot(ages_x,salary_y)
plt.plot(ages_x,salary2_y)
plt.legend(["fresher devloper","Senior developer"])
plt.show()
```

```python
plt.title("Salary median by ages")
plt.xlabel("Age")
plt.ylabel("Salary median")
plt.plot(ages_x,salary_y,label="Fresher Developer")
plt.plot(ages_x,salary2_y,label="Senior Developer")
plt.legend()
plt.show()
```

```python
plt.title("salary median by ages")
plt.xlabel("Age")
plt.ylabel("Salary median")
plt.plot(ages_x,salary_y,"k--",label="fresher")
plt.plot(ages_x,salary2_y,"b",label="senior")
plt.legend()
plt.show()
```

```python
plt.title("salary median")
plt.xlabel("Age")
plt.ylabel("Salary median")
plt.plot(ages_x,salary_y,color="k",linestyle="--",label="fresher")
plt.plot(ages_x,salary2_y,color="b",label="Senior")
plt.legend()
plt.show()

```

```python
plt.title("salary median")
plt.xlabel("Age")
plt.ylabel("Salary median")
plt.plot(ages_x,salary_y,color="r",linestyle="--",marker=".",label="fresher")
plt.plot(ages_x,salary2_y,color="b",marker=".",label="Senior")
plt.legend()
plt.show()

```

```python
salary3_y=[35000,38000,42000,45000,48000,50000,51000,52000,55000,60000,63000,65000]
plt.title("Salary median by ages")
plt.xlabel("Age")
plt.ylabel("Salary median")
plt.plot(ages_x,salary_y,color="r",linestyle="--",label="all dev")
plt.plot(ages_x,salary2_y,color="g",label="python dev")
plt.plot(ages_x,salary3_y,color="c",label="js dev")
plt.legend()
plt.show()
```

```python
plt.title("salary median by ages")
plt.xlabel("Age")
plt.ylabel("Salary median")
plt.plot(ages_x,salary_y,color="r",linestyle="--",label="all dev")
plt.plot(ages_x,salary2_y,color="c",label="python dev")
plt.plot(ages_x,salary3_y,color="c",label="js dev")
plt.legend()
plt.grid(True)
plt.show()

```

```python
plt.title("median salary by age")
plt.xlabel("Age")
plt.ylabel("Salary median")
plt.plot(ages_x,salary_y,color="r",linestyle="--",linewidth=3,label="All dev")
plt.plot(ages_x,salary2_y,color="b",linewidth=2,label="Python dev")
plt.plot(ages_x,salary3_y,color="c",label="js dev")
plt.legend()
plt.grid(True)
plt.show()
```

# Filling Area on lines Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("/storage/emulated/0/DS_PythonBasic/matplotlib5.csv")
age_x=df["age"]
py_salary=df["python"]
js_salary=df["js"]
#print(age_x)
#print(py_salary)
#print(js_salary)
              
```

```python

plt.plot(age_x,py_salary,label="python dev")
plt.plot(age_x,js_salary,label="js dev")

plt.legend()
plt.show()
```

```python
#add all dev salary
all_salary=[18000,19000,23000,25000,26000,28000,32000,34000,36000,37000,39000,42000]


plt.plot(age_x,all_salary,linestyle="--",label="all dev")
plt.plot(age_x,py_salary,label="python dev")
plt.plot(age_x,js_salary,label="js dev")
plt.legend()
plt.show()
```

```python
#fill area below the python salary
plt.plot(age_x,py_salary,label="python dev")
plt.plot(age_x,all_salary,linestyle="--",label="all dev")
plt.fill_between(age_x,py_salary)
plt.legend()
plt.show()

```

```python
#make fill color blure
plt.plot(age_x,py_salary,label="python dev")
plt.plot(age_x,js_salary,linestyle="--",label="js dev")
plt.fill_between(age_x,py_salary,alpha=0.25)


plt.legend()
plt.show()
```

```python
#the bydefault value for y2 is zero but we can change it
#i.e. if condition matches it will fill down otherwise it will fill up
overall_median=32000

plt.plot(age_x,py_salary,label="python dev")
plt.plot(age_x,js_salary,linestyle="--",label="js dev")
plt.fill_between(age_x,py_salary,overall_median,alpha=0.25)

plt.legend()
plt.show()

```

```python
plt.plot(age_x,py_salary,label="python dev")
plt.plot(age_x,js_salary, linestyle="--",label="js dev")

plt.fill_between(age_x,py_salary,
                 overall_median,
                 where=(py_salary>overall_median),
                 alpha=0.25)
plt.legend()
plt.show()
```

```python
#fill rigions correctly
plt.plot(age_x,py_salary,label="python dev")
plt.plot(age_x,js_salary,linestyle="--",label="js dev")
plt.fill_between(age_x,py_salary,
                 overall_median,
                 where=(py_salary>overall_median),
                 interpolate=True,alpha=0.25)

plt.legend()
plt.show()

```

```python
#fill up and down area with different colors

plt.plot(age_x,py_salary,label="python dev")
plt.plot(age_x,js_salary,linestyle="--",label="js dev")

plt.fill_between(age_x,py_salary,
                 overall_median,
                 where=(py_salary>overall_median),
                 interpolate=True,
                 alpha=0.25,label="above avg")
plt.fill_between(age_x,py_salary,
                overall_median,
                where=(py_salary<=overall_median),
                interpolate=True,
                color="red",
                alpha=0.25,label="below avg")
plt.legend()
plt.show()

```

```python
#fill area between two plots
plt.plot(age_x,py_salary,label="python dev")
plt.plot(age_x,js_salary,linestyle="--",label="js dev")

plt.fill_between(age_x,py_salary,
                 js_salary,
                 where=(py_salary>js_salary),
                 interpolate=True,
                 alpha=0.25,
                 label="above avg")
plt.fill_between(age_x,py_salary,
                 js_salary,
                 where=(py_salary<=js_salary),
                 interpolate=True,
                 alpha=0.25,
                 color="red",
                 label="below avg")
plt.legend()
plt.show()
                 


```

```python
#fill are between js dev salary and all dev salary
plt.plot(age_x,js_salary,label="js dev")
plt.plot(age_x,all_salary,linestyle="--",label="all dev")

plt.fill_between(age_x,js_salary,
                 all_salary,
                 where=(js_salary>all_salary),
                 interpolate=True,
                 alpha=0.25,
                 label="above avg")
plt.fill_between(age_x,js_salary,
                 all_salary,
                 where=(js_salary<=all_salary),
                 interpolate=True,
                 color="red",
                 alpha=0.25,
                 label="below avg")
plt.legend()
plt.show()

```

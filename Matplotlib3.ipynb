## Pie Charts in matplotlib

```python
from matplotlib import pyplot as plt
%matplotlib inline
slices=[60,40]

plt.pie(slices)

plt.title("My awesome pie chart")
plt.show()

```

```python
label=["part1","part2"]
plt.pie(slices,labels=label)

plt.title("my awesome pie chart")
plt.show()
```

```python
plt.pie(slices,labels=label,wedgeprops={"edgecolor":"black"})

plt.title("my awesome pie chart")
plt.show()

        


```

```python
slices2=[20,30,60,70,80]
label2=["part1","part2","part3","part4","part5"]
color=["red","green","yellow","blue","purple"]

plt.pie(slices2,labels=label2,colors=color,wedgeprops={"edgecolor":"black"})
plt.title("my awesome pie chart")
plt.show()

```

```python
#show percentage of each part
plt.pie(slices2,labels=label2,autopct="%1.1f%%",wedgeprops={"edgecolor":"black"})

plt.title("my awesome pie chart")

plt.show()

```

```python
#explode the piece of pie chart
expl=[0,0,0,0.1,0]
plt.pie(slices2,labels=label2,
        explode=expl,
        wedgeprops={"edgecolor":"black"},
       autopct="%1.1f%%")
plt.title("my awesome pie chart")
plt.show()
        
        
        
```

```python
#shadow of pie chart
plt.pie(slices2,labels=label2,
        autopct="%1.1f%%",
        wedgeprops={"edgecolor":"black"},
        shadow=True,
        explode=expl)
plt.show()
        
```

```python
#start angle means start the pie chart from that angle
plt.pie(slices2,labels=label2,
         startangle=90,autopct="%1.1f%%",
         wedgeprops={"edgecolor":"black"},
         shadow=True, explode=expl)
plt.title("my awesome pie chart")
plt.show()

```
